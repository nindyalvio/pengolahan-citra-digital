```python
%pylab notebook
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Deteksi Telapak Tangan
img = cv2.imread("/content/drive/MyDrive/Foto/palm.png")
scale_percent = 10 # percent of original size
width = int(img.shape[1] * scale_percent / 10)
height = int(img.shape[0] * scale_percent / 10)
dim = (width, height)

# resize image
img1 = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

print('Resized Dimensions : ',img1.shape)
print("Resized image")
cv2_imshow(img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mengubah Grayscale
gray = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
```

```python
# Deteksi Tepi Menggunakan Canny
edges = cv2.Canny(gray,40,55,apertureSize = 3)
print("edges in palm")
cv2_imshow(edges)
cv2.waitKey(0)
```

```python
edges = cv2.bitwise_not(edges)
print("edges in palm")
cv2_imshow(edges)
cv2.waitKey(0)
```

```python
cv2.imwrite("/content/drive/MyDrive/Foto/palmlines.png",edges)
palmlines = cv2.imread("/content/drive/MyDrive/Foto/palmlines.png")
img2 = cv2.addWeighted(palmlines, 0.3, img1, 0.7, 0)
cv2_imshow(img2)
cv2.waitKey(0)
```

```python
img = cv2.imread("/content/drive/MyDrive/Foto/kyungsoo.jpg")
scale_percent = 10 # percent of original size
width = int(img.shape[1] * scale_percent / 10)
height = int(img.shape[0] * scale_percent / 10)
dim = (width, height)

# resize image
img1 = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

print('Resized Dimensions : ',img1.shape)
gray_img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
print("Resized image")
cv2_imshow(img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
haar_face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt.xml")
```

```python
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5)
for (x, y, w, h) in faces:
  cv2.rectangle(img1, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
print("DO Kyungsoo")
cv2_imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY))
cv2.waitKey(0)
```

```python
# code untuk menampilkan citra
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# membaca gambar
sunrise = cv2.imread('/content/drive/MyDrive/Data Cuaca/Multi-class Weather Dataset/Sunrise/sunrise105.jpg')
# convert BGR ke RGB
sunrise = cv2.cvtColor(sunrise, cv2.COLOR_BGR2RGB)
# convert RGB ke HSV
hsv_sunrise = cv2.cvtColor(sunrise, cv2.COLOR_RGB2HSV)
# deklarasi batas bawah (orange cerah)
light_orange = (1,190,200)
# deklarasi batas atas (dark orange)
dark_orange = (18,255,255)
# tresholding
mask = cv2.inRange(hsv_sunrise, light_orange, dark_orange)
# impose gambar asli dengan mask
result = cv2.bitwise_and(sunrise, sunrise, mask=mask)
# plotting
plt.subplot(1,4,1)
plt.imshow(sunrise)
plt.subplot(1,4,2)
plt.imshow(hsv_sunrise)
plt.subplot(1,4,3)
plt.imshow(mask)
plt.subplot(1,4,4)
plt.imshow(result)
plt.show()
```
