```python
import cv2
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#tf.keras.applications.mobilenet.MobileNet
mobile = tf.keras.applications.mobilenet.MobileNet()
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import imagenet_utils
```

```python
citra = '/content/drive/MyDrive/Data Cuaca/Multi-class Weather Dataset/Rain/rain118.jpg'
img = image.load_img(citra)
```

```python
plt.imshow(img)
```

```python
resize_citra = image.img_to_array(img)
final_citra1 = np.expand_dims(resize_citra, axis=0)
final_citra = tf.keras.applications.mobilenet.preprocess_input(final_citra1)
```

```python
final_citra.shape
```

```python
predictions = mobile.predict(final_citra)
```

```python
hasil = imagenet_utils.decode_predictions(predictions)
```
